```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("stance_width_data5.csv")
```

```python
df['label_1']=df.label_1.str.lower()
```

```python
df
```

```python
df_filt = df.loc[~df.label_1.str.contains("not enough info") ]
```

```python
df_filt
```

```python
df_filt.loc[df_filt.label_1=='narrow','label_1']='01_narrow'
df_filt.loc[df_filt.label_1=='slightly narrow','label_1']='02_slightly_narrow'
df_filt.loc[df_filt.label_1=='ok','label_1']='03_ok'
df_filt.loc[df_filt.label_1=='slightly wide','label_1']='04_slightly_wide'
df_filt.loc[df_filt.label_1=='wide','label_1']='05_wide'
```

```python
df_filt2 =df_filt.query('confidence_min_wgt_avg > 0.7')
```

```python
fig,ax = plt.subplots(figsize=(10,5))
sns.scatterplot(x='label_1',y='stance_wgt_avg',data=df_filt2.sort_values('label_1') ,
                ax=ax, hue='confidence_min_wgt_avg')
```

```python
fig,ax = plt.subplots(figsize=(10,5))
svm =sns.scatterplot(x='label_1',y='stance_wgt_avg',data=df_filt2.sort_values('label_1') ,
                ax=ax, hue='confidence_min_wgt_avg')
figure=svm.get_figure()
figure.savefig('stance_label.pdf')
```
