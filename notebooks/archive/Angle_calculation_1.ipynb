```python
import numpy as np
import matplotlib.pyplot as plt
import math
import os
import pandas as pd
```

```python
path="/Users/gwattal/squatsmart/output_keypoints/"
entries = os.listdir(path)
```

```python
entries
```

```python
name=[]
angle_heel=[]
angle_ankle=[]
confidence_min_heel=[]
confidence_min_ankle=[]
for file in entries:
    print(file)
    x,y,c=np.loadtxt((path+file)).T
    r_shoulder_x=x[2]
    r_shoulder_y=y[2]
    r_shoulder_c=c[2]
    r_heel_x= x[24]
    r_heel_y= y[24]
    r_heel_c= c[24]
    r_ankle_x=x[11]
    r_ankle_y=y[11]
    r_ankle_c=c[11]
    l_shoulder_x=x[5]
    l_shoulder_y=y[5]
    l_shoulder_c=c[5]
    l_heel_x= x[21]
    l_heel_y= y[21]
    l_heel_c= c[21]
    l_ankle_x=x[14]
    l_ankle_y=y[14]
    l_ankle_c=c[14]
    ##
    #Angle calculation 
    theta_r=math.atan((r_shoulder_y-r_heel_y)/(r_shoulder_x -r_heel_x))
    theta_l=math.atan((l_shoulder_y-l_heel_y)/(l_shoulder_x -l_heel_x))
    ##
    angle_heel.append(180- theta_l - theta_r)
    confidence_min_heel.append(min(r_shoulder_c,r_heel_c,l_shoulder_c,l_heel_c))
    name.append(file.replace("_keypoints.txt",""))
    #Angle calculation 
    theta_r_2=math.atan((r_shoulder_y-r_ankle_y)/(r_shoulder_x -r_ankle_x))
    theta_l_2=math.atan((l_shoulder_y-l_ankle_y)/(l_shoulder_x -l_ankle_x))
    ##
    angle_ankle.append(180- theta_l_2 - theta_r_2)
    confidence_min_ankle.append(min(r_shoulder_c,r_ankle_c,l_shoulder_c,l_ankle_c))
df = pd.DataFrame({'name': name, 'angle_heel': angle_heel, "confidence_min_heel":confidence_min_heel,
                  'angle_ankle': angle_ankle, "confidence_min_ankle":confidence_min_ankle})
df=df.sort_values(by=['name'])
df.to_csv('shoulder_heel_angel_data4.csv')
print("Done")
```
