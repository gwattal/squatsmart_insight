## Data Preprocessing

In this notebook we are doing some exploratory work with our data.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import random
import pickle
```

```python
path_train="/home/ubuntu/side_output_keypoints/"
path_test="/home/ubuntu/side_output_keypoints/test_set/"
```

```python
features_train=[]
confidence_train=[]
y_train=[]
high=[37,38,39,40,41,42,45,48,51,72,75,76,78,79,80,82,83,88,89,91,92]
normal=[43,46,47,50,73,74,81,84,90,44,49,85,87]
files=os.listdir(path_train)
files2=[f for f in files if ".txt" in f ]
for file in files2:
    print(file)
    xi,yi,c=np.loadtxt(path_train+file).T
    features_train.append(np.hstack((xi,yi))/max(np.hstack((xi,yi))))
    confidence_train.append(c)
    movie_number=int(file[6:8])
    print(movie_number)
    if(movie_number in high):
        y_train.append("high")
    else:
        y_train.append("normal")
x_train=np.asarray(features_train)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='ovr')
model.fit(x_train, y_train)
# make predictions
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=6, random_state=0)
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
```
