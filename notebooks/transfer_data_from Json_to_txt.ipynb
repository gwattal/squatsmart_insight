```python
import json
import os
import numpy as np
```

```python
os.listdir()
```

```python
json_files = [pos_json for pos_json in os.listdir() if pos_json.endswith('.json')]


for json_file json_files:
    with open(json_file) as f:
        print(json_file)
        data = json.load(f)
        keypoints_list=data['people'][0]['pose_keypoints_2d']
        keypoints_array=np.asarray(keypoints_list).reshape(25,3)
        np.savetxt("../text-file/"+json_file.replace(".json",'.txt'),keypoints_array,delimiter=' ')
        print("done with",json_file)
print("done")
```
