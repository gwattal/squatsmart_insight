```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
```

```python
def FrameCapture(pathvideo): 

    # Path to video file 
    vidObj = cv2.VideoCapture(pathvideo) 

    # Used as counter variable 
    count = 1

    # checks whether frames were extracted 
    success = True

    while success: 
        # vidObj object calls read 
        # function extract frames 
        print(count)
        success, image = vidObj.read() 
        #print(image[0,0,:])
        # Saves the frames with frame-count
        
        #plt.imshow(image)
        matplotlib.image.imsave("/Users/gwattal/squatsmart/frames/MOV_0001_frame%d.png" %count, image[:,:,::-1]) 
        count += 1


# Calling the function
pathvideo="/Users/gwattal/squatsmart/videos/MOV_0001.mp4"
#pathframe="~/squatsmart/frames/"
FrameCapture(pathvideo) 

```
